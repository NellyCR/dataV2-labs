### Looking at the distributions of the data
Let's take a look at the distribution of the 4 stats that describe the physical characteristics of the players.

**Plot the four distributions about `height`, `weight`, `age` and `BMI`.**

```python
#your code here

import plotly.express as px

height = px.histogram(wnba["Height"])
weight = px.histogram(wnba["Weight"])
age = px.histogram(wnba["Age"])
BMI = px.histogram(wnba["BMI"])

```

```python
fig, [reb, ast, stl, pts, blk] = plt.subplots(1,5, figsize=(15,5))
reb.boxplot(wnba["REB"])
reb.set_title('REBOUNDS')
ast.boxplot(wnba["AST"])
ast.set_title('ASSIST')
stl.boxplot(wnba["STL"])
stl.set_title('STEALS')
pts.boxplot(wnba["PTS"])
pts.set_title('POINTS')
blk.boxplot(wnba["BLK"])
blk.set_title('BLOCKS')
plt.show()
```

**What conclusions do you think we can take from this plots?**

```python
#your conclusions here

# height = Hight is important, most of players are least 170cm and the most frequent hight is 185-189cm.

# weight = 75-89 is the most common weight which is  normal for relatively tall people.
# it is more common seeing players that weight less than 74kg than players that weight more than 90 kg.

# age = Most of player are in their 20's which is a young age, after 30 years are less common.
# The most common age is between 24-25.

# BMI= For most adults, an ideal BMI is in the 18.5 to 24.9 range , we can see that most of the players are between that range.
# Is really rare seeing a player with more than 26. 
```

In addition to what the describe() method already told us we can see that the physical characteristics of the players (apart from the age) more or less follow a standard distribution, which is expected when looking at the distribution of values that arise from a lot of factors that are independent from each other as is the case for many physical quantities. 

The height distribution looks like a bimodal. This may be due to the fact that players in basketball fall under two main categories (please note that this is a very gross generalization): shorter and more agile, and taller and less agile. Therefore there are less "average" height players since they will neither be as agile as the shorter players nor have the same impact in the paint (that is, under the basket) as a taller player.

The age distribution is a bit skewed to the right which is expected since most professional players stop playing after their prime physical years come to an end.

We could do the same for the main game stats. They are points, assists, blocks, rebounds and steals. 

**Now plot the distribution of the columns `REB`, `AST`, `STL`, `PTS` and `BLK` the same way you did in the last cell.**

```python
#your code here
import plotly.express as px

# Reb = the total number of rebounds (Off Reb + Def Reb)
# Ast = an assist is attributed to a player who passes the ball to a teammate in a way that leads to a score
# Stl = the number of steals by a defensive player or team
# Pts = the total number of points scored by field goals and free throws
# Blk = the number of blocks by a defensive player or team

REB = px.histogram(wnba["REB"])
AST = px.histogram(wnba["AST"])
STL = px.histogram(wnba["STL"])
PTS = px.histogram(wnba["PTS"])
BLK = px.histogram(wnba["BLK"])

```

**What conclusions do you think we can take from this plots?**

```python
#your conclusions here

# REBOUNDS / REB
# Rebounding is one of the most important phases of the game of basketball. 
# Rebounding gives a team possession of the basketball, and each possession helps both a team's offense and their defense and ultimately helps a team win basketball. games

# ASSIST / AST =

# STEALS / STL
# Players with high steal percentages can be a nice piece to have on defense.

# PTS
# We have almost the same amont of players that scored between 0-99 points (23 counts) or between 200-249 points (21 counts)

# BLK 
# A player with the ability to block shots can be a positive asset to a team's defense. 
# They can make it difficult for opposing players to shoot near the basket and by keeping the basketball in play.
# We see again 21 counts that have the most blocked (200-249) , are they the same 21 counts (players) scoring between 200-249 points ?
```

As expected all of the above distribution are heavily skewed to the right, since most players will have very low to average stats while there will be a handful of star players whose stats peak above everyone else. It is also important to think about the fact that we are simply taking the stats as they are without considering the minutes played by each player. Even though skill plays a very important factor in determining this kind of stats we also have to consider that players that play more minutes will, on average, score more points (or blocks, assists, etc.).

**For the sake of it let's look at the same distributions by dividing those stats by the minutes played for each player in the dataset.** 

```python
# I created new columns to dividing each stats by the minutes played

wnba['min_REB'] = wnba['REB'] / wnba['MIN']
wnba['min_AST'] = wnba['AST'] / wnba['MIN']                                     
wnba['min_STL'] = wnba['STL'] / wnba['MIN']
wnba['min_PTS'] = wnba['PTS'] / wnba['MIN']
wnba['min_BLK'] = wnba['BLK'] / wnba['MIN']

#  new data set to better visualize the new columns, and sort the stats by the most minutes played.

minutes = wnba[['MIN','min_REB','min_AST','min_STL', 'min_PTS' , 'min_PTS']]
minutes = minutes.sort_values(by='MIN', ascending=False) 

# To compare the stats of players with more  minutes played
minutes.head(25)

# To compare the stats of players with  less minutes played
#minutes.tail(25)

# We can also check the correlation
#minutes.corr()
```

```python
fig, [reb, ast, stl, pts, blk] = plt.subplots(1,5, figsize=(15,5))
reb.boxplot(wnba["REB"]/wnba['MIN'])
reb.set_title('REBOUNDS PER MIN')
ast.boxplot(wnba["AST"]/wnba['MIN'])
ast.set_title('ASSIST PER MIN')
stl.boxplot(wnba["STL"]/wnba['MIN'])
stl.set_title('STEALS PER MIN')
pts.boxplot(wnba["PTS"]/wnba['MIN'])
pts.set_title('POINTS PER MIN')
blk.boxplot(wnba["BLK"]/wnba['MIN'])
blk.set_title('BLOCKS PER MIN')
plt.show()
```
