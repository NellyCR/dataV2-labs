```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

canada = pd.read_csv('/Users/nelly/dataV2-labs/module-2/Explore-Correlations/CDD.csv', sep='\t')

```

```python
canada.head(50)
```

```python
canada2 = canada.corr()

canada2
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(canada2, linewidths=.5, annot=True, annot_kws={"size": 12}, cmap="YlGnBu")
```

```python
# High positive correlation 

canada['INSURANCE PAYMENTS'].corr(canada['EVACUATED'])

# For an insurance is hard to try to deny payments since an event like an evacuations is a national emergency. 
```

```python
# Negative correlation  

# With this correlation is so close to zero that is hard to draw a conclusion.
# I was expecting a higher negative correlation though, for me the more people evacuated the less fatalities. 

canada['FATALITIES'].corr(canada['EVACUATED'])
```

```python
# Correlation is almost zero

# The amount of people whose utility services (power, water, etc.) were interrupted/affected is not covered by
# the Disaster Financial Assistance Arrangements.

canada['UTILITY - PEOPLE AFFECTED'].corr(canada['FEDERAL DFAA PAYMENTS'])
```

```python

canada_one = canada[['INSURANCE PAYMENTS', 'EVACUATED']]

sns.pairplot(canada_one,)

# The correlation is .82 
```

```python
# Negative correlation is -0.003 (almost zero)

# On the top right plot, most of the ponits are distributed along the x axis , if we wanted to draw a line it would be
# slightly facing down so it confirms the small negative correlation. 


canada_two = canada[['FATALITIES','EVACUATED']]

sns.pairplot(canada_two)
```

```python
# Correlation almost zero : 0.022

# On the top bottom plot we can see that if we would draw a line most points will be really close to the x axis. 

canada_three = canada[['UTILITY - PEOPLE AFFECTED', 'FEDERAL DFAA PAYMENTS']]

sns.pairplot(canada_three)
```
