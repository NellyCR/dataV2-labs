```python
import requests


url = 'https://www.enforcementtracker.com/'
html = requests.get(url).content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
```

```python
table = soup.find_all('tr')

#rows= [row.text for row in table ]

#rows

#table

# td= table element

rows2= [row.find_all('td') for row in table ]

# you can aply .text to the elements when you use findall

rows3 = [ [ element.text for element in lst] for lst in rows2] 

rows3[1:]

```

```python
import pandas as pd

fines = pd.DataFrame(rows3[1:])
```

```python
fines.columns = ['Country','Authority','Date', 'Fine_EUR', 'Controller', 'GDPR_article', 'Type', 'Summary', 'Link']

fines.head(2)
```

```python
fines.drop(columns=['Link'], inplace=True)
```
